```python
import fitz  # PyMuPDF

pdf_path = file_path = "D:\Downloads\Subject-Independent_Continuous_Locomotion_Mode_Classification_for_Robotic_Hip_Exoskeleton_Applications.pdf"
doc = fitz.open(pdf_path)
doc
```

```python
text_all = []
doi = None
i = 0
print(type(doc), doc.page_count, doc.metadata)
print(doc[i].get_text("text"))
```

```python
# 获取页面的结构化文本信息（字典格式）
info = doc[i].get_text("dict")
info
```

```python
print(info.get("blocks", []))
len(info.get("blocks", []))
```

```python
spans = []
for b in info.get("blocks", []):
    if b.get("type", 0) != 0:
        continue
    for line in b.get("lines", []):
        # print(line)
        for s in line.get("spans", []):
            txt = (s.get("text") or "").strip()
            # print(txt)
            if not txt:
                continue
            y0 = float(s.get("bbox", [0, 0, 0, 0])[1])
            spans.append(
                dict(text=txt, size=float(s.get("size", 0)), y0=y0)
            )
            print(dict(text=txt, size=float(s.get("size", 0)), y0=y0))
```
